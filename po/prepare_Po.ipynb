```python
from PIL import Image, ImageFont, ImageDraw, ImageEnhance
import requests
from io import BytesIO
import json
import os
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from skimage.filters import threshold_otsu
from skimage.io import imread, imsave
from skimage.color import rgb2gray
from skimage.measure import label, regionprops
from skimage.transform import rescale, resize, downscale_local_mean
import random
import math
import imageio
import imgaug as ia
# from imgaug import augmenters as iaa
import imgaug.augmenters as iaa
from imgaug.augmentables.bbs import BoundingBox, BoundingBoxesOnImage
import glob
import random
import shutil
Image.MAX_IMAGE_PIXELS = None
```

```python
files = glob.glob('po-trophozoite/*.TIFF')
k = 367
for file in files:
    print(file)
    im = Image.open(file)
    W, H = im.size
    w = W + round(W*0.30)
    h = H + round(H*0.30)
#     print("{} {}".format(W, w))
    img = im.resize((w,h), Image.BILINEAR)
    print(img.size)
    a = img.crop((round(W*0.05), round(H*0.05), 1600 + round(W*0.05), 1200 + round(H*0.05)))
    a.save(os.path.join('tmp_images',"po-{:010d}.jpg".format(k)),"JPEG", quality=95, optimize=True, progressive=True)
    k += 1
#     c = crop(file,'tmp_images',1600,1200)
#     print(c)
```

```python
k = 511
files = glob.glob('po-schizont/*.TIFF')
for file in files:
    for n in range(6):
        img = Image.open(file)
        im = np.array(img)
        seq = iaa.Sequential([iaa.Affine(translate_percent={"x": (-0.25, 0.25), "y": (-0.25, 0.25)}), iaa.Crop(percent=(0.2), keep_size=True, sample_independently=False)])
        aug = seq(image=im)
        aug = Image.fromarray(aug)
        aug.save(os.path.join('tmp_images',"po-{:010d}.jpg".format(k)),"JPEG", quality=95, optimize=True, progressive=True)
        k += 1
# plt.imshow(aug)
```
